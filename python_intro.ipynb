## Conditional Statement/ Keputusan dengan adanya Kondisi

```python
# if-else-if
x = 10
y = 5
if x > y:
    print("x lebih besar dari y")
elif x < y:
    print("x lebih kecil dari y")
else:
    print("x sama dengan y")
```

```python
# nested-if
x = 10
y = 5
if x > y:
  print("x lebih besar dari y")
  if x > 0:
    print("x adalah bilangan positif")
  else:
    print("x adalah bilangan negatif")
elif x < y:
  print("x lebih kecil dari y")
else:
  print("x sama dengan y")
```

## Looping/ Perulangan

```python
# for loop
for i in range(5):
  print(i)
```

```python
# while do
x = 0
while x < 5:
  print(x)
  x += 1
```

```python
# do while
x = 0
while True:
  print(x)
  x += 1
  if x >= 5:
    break
```

## Costumer Call

```python
# Load Library
import pandas as pd
```

```python
# Buka Gdrive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Print Dataset

# Buat Folder di Google Drive [ MIKTI > Dataset > Filenya Dataset ]
path_dataset = "/content/drive/MyDrive/MIKTI/Dataset/Customer Call List.xlsx"
df = pd.read_excel(path_dataset)
df
```

```python
# Hapus data Duplikasi
df = df.drop_duplicates()
df
```

```python
# Drop Kolom
df = df.drop(columns=['Not_Useful_Column'])
df
```

```python
# - Last Name -
# Strip menghilangkan simbol gak sesuai
# df['Last_Name'] = df['Last_Name'].str.lstrip("/.")
# df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df['Last_Name'] = df['Last_Name'].fillna('')
df
```

```python
# - Phone Number -
df["Phone_Number"] = df["Phone_Number"].str.replace('N/a','')
df["Phone_Number"] = df["Phone_Number"].str.replace('|','')
df["Phone_Number"] = df["Phone_Number"].str.replace('/','')
df["Phone_Number"] = df["Phone_Number"].str.replace('-','')
df["Phone_Number"] = df["Phone_Number"].fillna('') # Data Kosong
df
```

```python
# Pecah Belah Data Adress menjadi 3 Bagian dengan (,)
df["Address"] = df["Address"].str.replace('N/a','')
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True) # Removed the extra positional argument. The expand argument is a keyword argument and should be passed as expand=True
df
```

```python
# - Do Not Contact -
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna('')
df
```

```python
# Paying Costumer
df.columns = df.columns.str.replace(' ', '_')
df["Paying_Customer"] = df["Paying_Customer"].str.replace('Yes','Y')
df["Paying_Customer"] = df["Paying_Customer"].str.replace('No','N')
df["Paying_Customer"] = df["Paying_Customer"].str.replace('N/a','')
df["Paying_Customer"] = df["Paying_Customer"].fillna('')
df
```

```python
# Filtering Data yang gak mau di Kontak (Y)
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
# Filtering data yang gak punya Nomor HP/ Data Kosong
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df
```

```python
# Filtering Data Costumer yang bayar (Y)
for x in df.index:
    if df.loc[x, "Paying_Customer"] == 'N':
        df.drop(x, inplace=True)

df
```

```python
df['State'] = df['State'].apply(lambda x: '-' if x is None else x)
df['Zip_Code'] = df['Zip_Code'].apply(lambda x: '-' if x is None else x)
df
```

```python
# reset urutan kolom menjadi terurut
df = df.reset_index(drop=True)
df
```

```python
# Export to Excel
df.to_excel('cleaned_customer_list.xlsx', index=False)

# Download the Excel file
from google.colab import files
files.download('cleaned_customer_list.xlsx')
```
